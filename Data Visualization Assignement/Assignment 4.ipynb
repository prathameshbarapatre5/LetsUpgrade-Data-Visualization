```
#Importing Libraries 
import numpy as np
import pandas as pd
import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```
#Q3) plot a Pie Chart of Indian Car companies market share % values,max ( 6 Companies market share )

#Indian Companies and their share in %
maruti = 50
hyundai = 17.4
tata = 7
kia = 5.8
mahindra = 5.6
renault = 3.3
toyota = 3.1
honda = 2.9

y = np.array([50,17.4,7,5.8,5.6,3.3,3.1,2.9])
names = ["maruti","hyundai","tata","kia","mahindra","renault","toyota","honda"]

#plotting pie chart
df = pd.DataFrame(y)
df.rename(columns={0:"Market_Shares"},inplace = True)
df['Company_Names'] = names
```

```
df
```

```
fig = px.pie(df,values="Market_Shares", names = "Company_Names")
fig.show()
```
