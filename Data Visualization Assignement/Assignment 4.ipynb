```python
#Q5 Plot a Line Chat with India's active covid-19 case recovery (maximum 12 Months)

#Importing Dataset 
df = pd.read_csv("E://Desktop folder//Internship Projects//LetsUpgrade//Data Visualization//Data Visualization Assignement//covid_19_india.csv")
df.head()
```

```python
df.drop(df.columns[[0,2,3,4,5]],inplace = True,axis = 1)
df.head()
```

```python
df['Date']= pd.to_datetime(df['Date']) #Date is converted to DateTime Format          
df = df[df['Date'].dt.year==2020] 
```

```python
df['month'] = df['Date'].dt.month
df.head()
```

```python
df_cured = df['Cured'].groupby(df['month']).sum()
df_deaths = df['Deaths'].groupby(df['month']).sum()
df_confirmed = df['Confirmed'].groupby(df['month']).sum()
df_Month = df['month'].groupby(df['month']).sum()
```

```python
df1 = df_cured.sum()
df2 = df_deaths.sum()
df3 = df_confirmed.sum()
df4 = df_Month.sum()

df1= pd.DataFrame(df_cured)
df2= pd.DataFrame(df_deaths)
df3= pd.DataFrame(df_confirmed)
df4= pd.DataFrame(df_Month)

cols = [df1,df2,df3,df4]
df.total = pd.concat(cols,axis=1)
```

```python
df.total
```

```python
df['month'] = df['month']/1000
plt.figure(figsize=(10,10))
plt.title("Covid Cured")
sns.lineplot(x='month',y='Cured',data=df.total,color="b",lw=3,marker='o',markersize=10)
```
